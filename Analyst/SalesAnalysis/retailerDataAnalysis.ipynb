```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#from sklearn.cross_validation import train_test_split
import os
import glob
import re
import random
import datetime
```

```python
# read all csv files present in folders and subfolders

absPath = os.getcwd()

files = []
for r, d, f in os.walk(absPath):
    for file in f:
        if '.csv' in file:
            files.append(os.path.join(r, file))
```

```python
# merging all datasets into one

lis = []
for filename in files:
    df = pd.read_csv(filename, index_col=None, header=0)
    lis.append(df)

mainframe = pd.concat(lis, axis=0, ignore_index=True)
```

```python
def getNumericData(data):
    newData = []
    for strng in data:
        try:
            result = re.sub('[^0-9]','', strng)
        except:
            result = 0
        newData.append(int(result))
    return(newData)
```

```python
def seperateStoreWiseData(data):
    if  data['Store Code'] == 1:
        lis1.append(data)
    elif data['Store Code'] == 2:
        lis2.append(data)
    elif data['Store Code'] == 3:
        lis3.append(data)
    elif data['Store Code'] == 4:
        lis4.append(data)
    elif data['Store Code'] == 5:
        lis5.append(data)
    else:
        lis6.append(data)
    return
```

```python
def getTotalUniqueValue(col):
    return(col.nunique())
```

```python
def getMostFrequentValue(col):
    return(col.value_counts().idxmax())
```

```python
#PREPROCESSING
# clean alpha numeric dimensions
mainframe['SKU Code'] = getNumericData(mainframe['SKU Code'])
mainframe['Brand Code'] = getNumericData(mainframe['Brand Code'])
mainframe['Category'] = getNumericData(mainframe['Category'])
mainframe['Bill Number'] = getNumericData(mainframe['Bill Number'])
mainframe['Store Code'] = getNumericData(mainframe['Store Code'])
```

```python
# rectify sales price
# Clean brand code, catergory
for index, row in mainframe.iterrows():
    # clean, mrp sales price conflict
    if row['MRP'] < row['Sales Price']:
        mainframe.set_value(index, 'Sales Price', row['MRP'])
    # handle missing values
    if row['Brand Code']== 0:
        if row['Category'] == 0:
            _a = random.randint(0, 11)
            mainframe.set_value(index, 'Brand Code', _a)
            mainframe.set_value(index, 'Category', _a)
        else:
            mainframe.set_value(index, 'Brand Code', row['Category'])
```

```python
# drop rows with 0 mrp
mainframe.drop(mainframe[mainframe['Sales Price']== 0].index, inplace = True)
        
# drop duplicates
mainframe.drop_duplicates(['SKU Code', 'Brand Code', 'Category', 'Bill Number', 'Store Code', 'Sale Date', 'MRP', 'Sales Price', 'Sales Qty'], keep='first', inplace=True)
```

```python
mainframe.head()
```

```python
# add new features
#discount feature

# get week nos
weekLis = []
monthLis = []
_strDate = mainframe['Sale Date']
for _str in _strDate:
    _abc = [int(_a) for _a in _str.split('-')]
    _x = datetime.datetime(_abc[0], _abc[1], _abc[2])
    if _abc[0] != 2017:
        _mnth = _abc[1] + (12 *(_abc[0]-2017))
        _week = int(_x.strftime("%U")) + int(datetime.date(2017, 12, 31).isocalendar()[1])
    else:
        _mnth = _abc[1]
        _week = _x.strftime("%U")
    weekLis.append(_week)
    monthLis.append(_mnth)

mainframe['Sale Week'] = weekLis
mainframe['Sale Month'] = monthLis
mainframe['Discount in %'] = ((mainframe['MRP'] - mainframe['Sales Price'])*100)/ mainframe['MRP']
mainframe['Total Sales Amount'] = mainframe['Sales Price'] * mainframe['Sales Qty']
mainframe['Total MRP Amount'] = mainframe['MRP'] * mainframe['Sales Qty']
```

```python
# Drop irrelevant columns
mainframe.drop(['Sale/Return', 'Sale Date', 'Bill Number'], axis=1, inplace= True)
```

```python
# arrange dataset
mainframe = mainframe[['SKU Code', 'Brand Code', 'Category', 'Store Code', 'Sale Week', 'Sale Month','MRP', 'Discount in %', 'Sales Qty', 'Total MRP Amount', 'Total Sales Amount']]
```

```python
mainframe.head()
```

```python
# dictionary to map brand code to corresponding category
brandToCategory = {}
for index, row in mainframe.iterrows():
    key = row['Brand Code']
    value = row['Category']
    if key not in brandToCategory.keys():
        brandToCategory[key] = value
```

```python
# Monthly Sales Report Data

monthlySalesReport = (mainframe.groupby(['Store Code', 'Sale Month'], as_index=False).agg({'SKU Code' : getTotalUniqueValue, 
                                                                                            'Brand Code': getMostFrequentValue,
                                                                                            'Category': getMostFrequentValue,
                                                                                            'MRP': 'sum',
                                                                                            'Sales Qty': 'sum',
                                                                                            'Total MRP Amount': 'sum',
                                                                                            'Total Sales Amount':'sum'}).rename(columns={'SKU Code':'no. of unique_SKU_code',
                                                                                                                                           'Brand Code':'Most_Sold_Brand',
                                                                                                                                           'Category':'Most_Sold_Category',
                                                                                                                                           'MRP':'Total MRP',
                                                                                                                                           'Sales Qty':'No. Of Items Sold',
                                                                                                                                           'Total MRP Amount':'MRP Sum',
                                                                                                                                            'Total Sales Amount':'Total Sales Amount'}))
```

```python
monthlySalesReport.head()
```

```python
monthlySalesReport['Total Discount'] = ((monthlySalesReport['MRP Sum'] - monthlySalesReport['Total Sales Amount'])*100)/ monthlySalesReport['MRP Sum']
# arrange dataset
monthlySalesReport = monthlySalesReport[['Store Code', 'Sale Month','no. of unique_SKU_code', 'Most_Sold_Brand', 'Most_Sold_Category', 'Total MRP', 'Total Discount', 'MRP Sum', 'Total Sales Amount']]
```

```python
productSellReportMonthly = (mainframe.groupby(['Store Code', 'Sale Month', 'Brand Code'], as_index=False).agg({'Sales Qty':'sum',
                                                                                                        'Total Sales Amount':'sum',
                                                                                                        'Sale Week': getMostFrequentValue}).rename(columns={'Sales Qty':'Quantity Sold',
                                                                                                                                                            'Total Sales Amount':'Total Product Sale',
                                                                                                                                                            'Sale Week':'max Sale in Week(Year)'}))
```

```python
#productSellReportMonthly['Category'] = brandToCategory[productSellReportMonthly['Brand Code']]
# insert category in dataset
cate = []
for index, row in productSellReportMonthly.iterrows():
    value = brandToCategory[row['Brand Code']]
    cate.append(value)
productSellReportMonthly['Category'] = cate
```

```python
#Product sell report montly: monthly average sell of product store wise with week of year having maximum sale of particular product
productSellReportMonthly = productSellReportMonthly[['Store Code', 'Sale Month', 'Brand Code', 'Category','Quantity Sold', 'Total Product Sale', 'max Sale in Week(Year)']]
```

```python
productSellReportMonthly.head()
```

```python
# weekly sale report
productSellReportWeekly = (mainframe.groupby(['Store Code', 'Sale Week', 'Brand Code'], as_index=False).agg({'Sales Qty':'sum',
                                                                                                            'Total Sales Amount':'sum'}).rename(columns={'Sales Qty':'Quantity Sold',
                                                                                                                                                        'Total Sales Amount':'Weekly Sales'}))
```

```python
#productSellReportWeekly['Category'] = brandToCategory[productSellReportWeekly['Brand Code']]
# insert category in dataset
cate = []
for index, row in productSellReportWeekly.iterrows():
    value = brandToCategory[row['Brand Code']]
    cate.append(value)
productSellReportWeekly['Category'] = cate
```

```python
#Product sell report weekly: weekly sell of product store wise 
productSellReportWeekly = productSellReportWeekly[['Store Code', 'Sale Week', 'Brand Code', 'Category','Quantity Sold', 'Weekly Sales']]
```
